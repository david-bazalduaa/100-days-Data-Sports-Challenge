```python
import pandas as pd
from statsbombpy import sb
import os

# Define the local directory path for saving raw data
data_dir = "/Users/davidbazalduamendez/Documents/GitHub/100-days-Data-Sports-Challenge/data/raw/statsbomb/"

# Create the directory safely; ignores error if it already exists
os.makedirs(data_dir, exist_ok=True)
```

```python
# Review the ids 
sb.competitions()
# Champions League id 16
```

```python
competitions_df = sb.competitions()

# Filter the DataFrame to show only rows where the competition name is 'La Liga'
la_liga_df = competitions_df[competitions_df['competition_name'] == 'La Liga']

display(la_liga_df)
# la liga id = 11
```

```python
# Fetch matches for La Liga (11) in the 2020/2021 season (90)
matches_df = sb.matches(competition_id=11, season_id=90)

# Display the first few rows to explore the match schema
display(matches_df.head())
```

```python
# Extract the match_id from the first row
first_match_id = matches_df.iloc[0]['match_id']

# Fetch the event data for this specific match
events_df = sb.events(match_id=first_match_id)

# View the columns and the first few rows
display(events_df.head())
```

```python
events_df['type'].unique()
```

```python
# Filtrar el DataFrame original para quedarnos solo con los Tiros
shots_df = events_df[events_df['type'] == 'Shot']

# Extraer los nombres de las columnas que contienen detalles específicos del tiro
shot_columns = [col for col in shots_df.columns if col.startswith('shot_')]

# Mostrar el jugador, el minuto y todos los detalles del tiro
display(shots_df[['player', 'minute', 'location'] + shot_columns].dropna(axis=1, how='all').head())
```

```python
def download_match_events_individually(comp_id, season_id, comp_name, season_name):
    matches_df = sb.matches(competition_id=comp_id, season_id=season_id)
    
    # Create a specific folder for this season to keep things organized
    safe_comp = str(comp_name).replace(" ", "_")
    safe_season = str(season_name).replace("/", "_")
    save_dir = f"data/raw/statsbomb/{safe_comp}_{safe_season}/"
    os.makedirs(save_dir, exist_ok=True)
    
    for index, row in matches_df.iterrows():
        match_id = row['match_id']
        file_path = os.path.join(save_dir, f"{match_id}.csv")
        
        # Check if file already exists to avoid re-downloading
        if not os.path.exists(file_path):
            try:
                match_events = sb.events(match_id=match_id)
                match_events.to_csv(file_path, index=False)
                print(f"Saved match {match_id}")
            except Exception as e:
                print(f"Error downloading match {match_id}: {e}")
        else:
            print(f"Match {match_id} already exists. Skipping.")

# download_match_events_individually(11, 90, "La Liga", "2020_2021")
```
